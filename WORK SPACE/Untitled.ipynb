```python
import numpy as np
import pandas as pd
from datetime import datetime as dt
```

```python
file_name = 'vn_express.tsv'
```

```python
df = pd.read_csv(file_name, '\t')
df
```

# Preprocess Date Column

```python
temp_df = df['Date'].str.split(', ')
df['Date'] = temp_df.str[1]
df['Publish Time'] = temp_df.str[2].str.split(' ').str[0]
```

```python
df['Date'] = pd.to_datetime(df['Date'], format='%d/%m/%Y')
```

```python
df['Publish Year'] = df['Date'].dt.year
```
